# Answering the question of why there is a small lag at the beginning

* Hypothesis: because they need to get past the initial treshold of the minimum amount of viable genotypes that have to be present to allow a full viable path to exist

```python
# Parameters
nsim = 100
sites = 8
fraction_viable = 0.50

# Simulating a single gpmap as template, same for all. Generate once.
genotypes = generate_genotypes('AB',sites)

# Both the source and target must be viable
phenotypes = list(np.random.choice([1,0], 
                                   size = len(genotypes)-2, 
                                   p = [fraction_viable, 1-fraction_viable]))
phenotypes.insert(0, 1)
phenotypes.insert(len(genotypes)+1, 1)

gpm = gpmap.GenotypePhenotypeMap(wildtype = genotypes[0],
                                 genotypes = genotypes,
                                 phenotypes = phenotypes)
```

```python
gpg = gpgraph.GenotypePhenotypeGraph(gpm)
```

```python
total_nodes = 2**sites
```

```python
# Minimum number of nodes in the middle should be 256 - 2 (source and target).
# If there is less than that, a path cannot exist
```

```python
minimum_nodes = sites + 1
```

```python
# How the percentage of viable genotypes influences the number of avaible nodes as size of map changes
x_data = np.arange(0, 1, 0.05)
y_data = np.zeros(x_data.size)
for counter, value in enumerate(x_data):
    y_data[counter] = value * total_nodes

y_data    
```

```python
minimum_nodes_line = [minimum_nodes for i in x_data]
total_nodes_line = [total_nodes for i in x_data]
vertical_line = [0.06 for i in y_data]
```

```python
plt.figure(figsize =(10,7))
plt.plot(x_data, y_data, 'ok-', label = "Available nodes for an eight site map")
plt.plot(x_data, minimum_nodes_line, 'r--', label = "Minimum amount of nodes to connect a path")
plt.plot(x_data, total_nodes_line, '-.')
plt.plot(vertical_line, y_data, label = "y = 0.06")
plt.xlabel('Fraction of viable genotypes')
plt.ylabel('Nodes present')
plt.legend()
plt.savefig("data/3.svg")
```

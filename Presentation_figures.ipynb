```python
import gpmap
import gpgraph
import gpvolve
import itertools
import random
import numpy as np
import math
import gzip, pickle, pickletools
import networkx as nx
import copy
from gpgraph import get_neighbors
import gpvolve

%matplotlib inline
from matplotlib import pyplot as plt

def generate_genotypes(alphabet,L):
    data = [''.join(v) for v in list(itertools.product(alphabet,repeat=L))]
    return data
```

```python
import itertools
def itertools_version(alphabet,L):
    return list(itertools.product(alphabet,repeat=L))
data = itertools_version("AB",4)

data = [''.join(v) for v in data]
fraction_viable = 0.4
# Parameters
genotypes = data
wildtype = data[0]
target = data[-1]
phenotypes = list(np.random.choice([1,0], 
                                   size = len(genotypes)-2, 
                                   p = [fraction_viable, 1-fraction_viable]))
phenotypes.insert(0, 1)
phenotypes.insert(len(genotypes)+1, 1)


# Simulate the genotype phenotype map
gpm = gpmap.GenotypePhenotypeMap(wildtype, genotypes, phenotypes = phenotypes)

G = gpgraph.GenotypePhenotypeGraph(gpm)
G.add_model(gpgraph.strong_selection_weak_mutation)

# Draw the graph
figure = gpgraph.draw_gpgraph(G,
                              node_size = 400,
                              edge_arrows = False,
                              edge_colors='gray',
                              node_linewidths=1,
                              figsize=(6,4),
                              cmap = 'Purples',
                              colorbar = True,
                              cmap_max = 0.95,
                              cmap_min = 0.20)

plt.savefig("data/4.svg")
```

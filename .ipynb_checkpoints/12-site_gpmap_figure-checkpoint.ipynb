# Drawing a 12-site GP graph

```python
import gpmap
import gpgraph
import gpvolve
import itertools
import random
import numpy as np
import math
import gzip, pickle, pickletools
import networkx as nx
import copy
from gpgraph import get_neighbors
from matplotlib import pyplot as plt

def generate_genotypes(alphabet,L):
    data = [''.join(v) for v in list(itertools.product(alphabet,repeat=L))]
    return data
```

```python
# Read the file
data = []
with open('data/twelve_sites.txt', 'r') as file:
     for counter, item in enumerate(file):
        data.append(item.rstrip('\n'))
```

```python
# Parameters
genotypes = data
wildtype = data[0]
target = data[-1]
# phenotypes = np.random.rand(len(genotypes))
phenotypes = [counter/len(genotypes) for counter, item in enumerate(genotypes)]

# Simulate the genotype phenotype map
gpm = gpmap.GenotypePhenotypeMap(wildtype, genotypes, phenotypes = phenotypes)
```
